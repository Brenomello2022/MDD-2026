2 O Conceito de Fato (Granularidade)

Na ciência de dados, chamamos de Fato a ocorrência de um evento que pode ser quantificado e observado em um momento específico do tempo e do espaço.

Cada linha é um Fato: No dataset da PRF, cada linha representa um "acidente/causa". Granularidade: É o nível de detalhe do fato.

Alta Granularidade: Dados detalhados (ex: cada pessoa envolvida no acidente).

Baixa Granularidade: Dados agregados (ex: total de acidentes por mês no Paraná).

3 Medidas (Quantitativas) vs. Dimensões (Qualitativas)

Para analisar um fato, dividimos as colunas em dois grupos:

A) Medidas (Fatos Numéricos) São os dados que podemos somar, tirar média ou contar. Respondem à pergunta: "Quanto?"

B) Dimensões (Contexto) São os dados que categorizam o fato. Respondem às perguntas: "Onde?", "Quando?", "Como?" e "Quem?"
