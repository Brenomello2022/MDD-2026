```python
## 5) Descobrindo Tipos de Dados com type()

Como o Python é uma linguagem de **tipagem dinâmica**, às vezes precisamos confirmar que tipo de dado está guardado em uma variável.

**Sintaxe:** `type(variavel)`

**Exemplos Comuns:**
- `<class 'int'>` para números inteiros.
- `<class 'float'>` para números com casas decimais.
- `<class 'str'>` para textos (strings).
- `<class 'bool'>` para valores lógicos (True/False).
```

```python
numero_inteiro = 42
numero_decimal = 3.14
texto = "Olá, Python!"
valor_logico = True

print(f"Tipo de {numero_inteiro}: {type(numero_inteiro)}")
print(f"Tipo de {numero_decimal}: {type(numero_decimal)}")
print(f"Tipo de '{texto}': {type(texto)}")
print(f"Tipo de {valor_logico}: {type(valor_logico)}")
```

```python
## 6) Conversão de Tipos (Casting)

Muitas vezes recebemos um dado como texto (ex: vindo de um formulário) e precisamos transformá-lo em número para fazer cálculos.

- **int()**: Converte para inteiro. Ex: `int("10")` vira `10`.
- **float()**: Converte para número decimal. Ex: `float(5)` vira `5.0`.
- **str()**: Converte qualquer valor em texto. Ex: `str(100)` vira `"100"`.

**Dica:** O casting é essencial para evitar erros de tipo ao concatenar textos com números!
```

```python
idade_texto = "25"
idade_numero = int(idade_texto)
print(f"Idade como texto: '{idade_texto}' (tipo: {type(idade_texto)})")
print(f"Idade como número: {idade_numero} (tipo: {type(idade_numero)})")


# Número para string (útil para concatenação)
ano = 2026
mensagem = "Estamos em " + str(ano)
print(mensagem)
```

```python
## 7) Operadores Matemáticos Básicos

O Python funciona como uma calculadora poderosa e intuitiva.

- `+` (Adição) e `-` (Subtração)
- `*` (Multiplicação) e `/` (Divisão - sempre retorna um float)
- `//` (Divisão Inteira - descarta as casas decimais)
- `%` (Módulo - retorna o resto da divisão)
- `**` (Exponenciação - elevar um número à potência de outro)
```

```python
print("Operadores básicos:")
print(f"{a} + {b} = {a + b}")  # Adição
print(f"{a} - {b} = {a - b}")  # Subtração
print(f"{a} * {b} = {a * b}")  # Multiplicação
print(f"{a} / {b} = {a / b}")  # Divisão (retorna float)

print("\nOperadores especiais:")
print(f"{a} // {b} = {a // b}")  # Divisão inteira (descarta decimais)
print(f"{a} % {b} = {a % b}")    # Módulo (resto da divisão)
print(f"{a} ** {b} = {a ** b}")  # Exponenciação (10 elevado a 3)
```

```python
## 8) Operadores Lógicos

São usados para combinar múltiplas condições em uma única expressão, retornando sempre `True` ou `False`.

- **and** (e): Retorna verdadeiro apenas se **todas** as condições forem verdadeiras.
- **or** (ou): Retorna verdadeiro se **pelo menos uma** das condições for verdadeira.
- **not** (não): Inverte o valor lógico. Se algo é `True`, o `not` transforma em `False`.

**Exemplo:** `(idade >= 18) and (tem_carteira == True)`
```

```python
idade = 20
tem_carteira = True
tem_carro = False

print("Operador AND (e):")
print(f"Idade >= 18? {idade >= 18}")
print(f"Tem carteira? {tem_carteira}")
print(f"Pode dirigir? {(idade >= 18) and tem_carteira}")

print("\nOperador OR (ou):")
print(f"Tem carteira? {tem_carteira}")
print(f"Tem carro? {tem_carro}")
print(f"Tem pelo menos um? {tem_carteira or tem_carro}")

print("\nOperador NOT (não):")
print(f"Tem carro? {tem_carro}")
print(f"NÃO tem carro? {not tem_carro}")

print("\nCombinando operadores:")
pode_viajar = (idade >= 18) and (tem_carteira or tem_carro)
print(f"Pode viajar? {pode_viajar}")
```

```python
## 9) Listas (Mutáveis)

As **Listas** são coleções ordenadas de itens que podem ser alterados (mutáveis). No Python, usamos colchetes `[]` para criá-las. No Notebook, elas são exibidas de forma muito organizada.

- **Sintaxe:** `minha_lista = [1, 2, "texto", True]`
- **Acesso por Índice:** O Python começa a contar do 0. Então, `minha_lista[0]` é o primeiro item.
- **Mutabilidade:** Podes mudar um valor específico da lista a qualquer momento.
```

```python
# Criando e manipulando listas

# Criando uma lista
frutas = ["maçã", "banana", "laranja", "uva"]
print(f"Lista original: {frutas}")

print(f"\nPrimeira fruta (índice 0): {frutas[0]}")
print(f"Última fruta (índice -1): {frutas[-1]}")

# Modificando um elemento (mutabilidade)
frutas[1] = "morango"
print(f"\nApós trocar banana por morango: {frutas}")

# Adicionando elementos
frutas.append("abacaxi")
print(f"\nApós adicionar abacaxi: {frutas}")

# Removendo elementos
frutas.remove("laranja")
print(f"\nApós remover laranja: {frutas}")

# Tamanho da lista
print(f"\nNúmero de frutas: {len(frutas)}")
```

```python
## 10) Tuplas (Imutáveis)

As **Tuplas** são muito parecidas com as listas, mas com uma diferença crucial: elas são **imutáveis**. Uma vez criadas, não podes alterar, adicionar ou remover itens. Usamos parênteses `()` para criá-las.

- **Sintaxe:** `minha_tupla = (10, 20, 30)`
- **Por que usar?** São mais rápidas que as listas e garantem que dados sensíveis (como coordenadas geográficas ou configurações) não sejam alterados acidentalmente pelo código.
```

```python
# Criando e usando tuplas

# Exemplo: coordenadas geográficas (não devem mudar)
coordenadas = (-23.5505, -46.6333)  # São Paulo
print(f"Coordenadas de São Paulo: {coordenadas}")

# Acessando elementos (igual às listas)
print(f"Latitude: {coordenadas[0]}")
print(f"Longitude: {coordenadas[1]}")

# Tuplas também podem ter tipos diferentes
pessoa = ("João", 30, True, 1.75)
nome, idade, ativo, altura = pessoa  # Desempacotamento
print(f"\nNome: {nome}, Idade: {idade}, Ativo: {ativo}, Altura: {altura}m")

# Tentativa de modificação (vai gerar erro se descomentar)
# coordenadas[0] = -22.0  # TypeError: 'tuple' object does not support item assignment

print(f"\nTipo de coordenadas: {type(coordenadas)}")
```

```python
## 11) Estruturas Condicionais (if, elif, else)

As estruturas condicionais permitem que o código siga caminhos diferentes dependendo de uma condição lógica.

**Importante:** Em Python, a **indentação** (o espaço no início da linha) é obrigatória. É ela que diz ao Python o que está "dentro" do seu `if`.

- **if:** "Se" a condição for verdadeira, faça isso.
- **elif:** "Senão, se" a condição anterior foi falsa, tente esta.
- **else:** "Senão", se nada acima funcionou, faça isso por padrão.
```

```python
# Exemplo 1: if simples

idade = 20

if idade >= 18:
    print("Você é maior de idade!")
    print("Pode votar e dirigir.")

print("Esta linha está fora do if")
```

```python
# Exemplo 2: if, elif, else

nota = 7.5

if nota >= 9:
    print("Excelente! Nota A")
elif nota >= 7:
    print("Bom! Nota B")
elif nota >= 5:
    print("Regular. Nota C")
else:
    print("Precisa melhorar. Nota D")

print(f"\nSua nota foi: {nota}")
```

```python
# Exemplo 3: Combinando condições com operadores lógicos

usuario = "admin"
senha = "1234"

if usuario == "admin" and senha == "1234":
    print("✓ Acesso permitido!")
    print("Bem-vindo, administrador!")
else:
    print("✗ Acesso negado!")
    print("Usuário ou senha incorretos.")
```

```python
## 12) Estruturas de Repetição (for e while)

As estruturas de repetição permitem que você execute o mesmo bloco de código várias vezes, poupando linhas de programação.

### for (para)
É usado quando você sabe (ou o Python sabe) exatamente quantas vezes repetir.
- **Percorrer listas:** Ele pega um item por vez da lista até chegar ao último.
- **Função range():** Cria uma sequência numérica (ex: `range(5)` gera de 0 a 4) para repetições controladas.

### while (enquanto)
É usado quando você não sabe exatamente quantas vezes o código vai rodar. Ele continuará repetindo enquanto uma condição for verdadeira.

⚠️ **Cuidado com o Loop Infinito:** Se a condição do `while` nunca se tornar falsa, o código nunca para de rodar. É necessário garantir que algo mude dentro do bloco para encerrar a repetição.
```

```python
# Exemplo 1: for com lista

frutas = ["maçã", "banana", "laranja"]

print("Frutas disponíveis:")
for fruta in frutas:
    print(f"- {fruta}")
```

```python
# Exemplo 2: for com range()

print("Contando de 0 a 4:")
for i in range(5):
    print(i)

print("\nContando de 1 a 10:")
for i in range(1, 11):
    print(i, end=" ")

print("\n\nContando de 2 em 2:")
for i in range(0, 11, 2):
    print(i, end=" ")
```

```python
# Exemplo 3: while

contador = 1

print("Contagem regressiva:")
while contador <= 5:
    print(f"Número: {contador}")
    contador += 1  # Importante! Sem isso vira loop infinito

print("Fim da contagem!")
```

```python
# Exemplo 4: while com break

print("Procurando um número específico:")
numero = 1

while True:  # Loop infinito proposital
    if numero == 7:
        print(f"Encontrei o número {numero}!")
        break  # Para o loop
    print(f"Número atual: {numero}")
    numero += 1

print("Loop encerrado.")
```

```python
## 13) Importando Bibliotecas (import)

Uma **biblioteca** é um conjunto de códigos escritos por outras pessoas que você pode usar no seu projeto para não precisar "reinventar a roda". A biblioteca `random`, por exemplo, serve para gerar números aleatórios ou fazer escolhas randômicas.

- **import:** Carrega a biblioteca inteira para o seu Notebook.
- **Uso:** Após importar, você acessa as funções usando o ponto (Ex: `biblioteca.funcao()`).

Algumas bibliotecas populares:
- `random` - números aleatórios
- `math` - funções matemáticas avançadas
- `datetime` - trabalhar com datas e horas
```

```python
# Exemplo 1: Biblioteca random

import random

# Gerar número aleatório entre 1 e 10
numero_aleatorio = random.randint(1, 10)
print(f"Número aleatório: {numero_aleatorio}")

# Escolher elemento aleatório de uma lista
cores = ["vermelho", "azul", "verde", "amarelo"]
cor_sorteada = random.choice(cores)
print(f"Cor sorteada: {cor_sorteada}")

# Embaralhar uma lista
numeros = [1, 2, 3, 4, 5]
random.shuffle(numeros)
print(f"Lista embaralhada: {numeros}")
```

```python
## 14) Funções (def)

Uma **função** é um bloco de código que só é executado quando é chamado. Elas servem para evitar a repetição de código e para organizar a lógica em partes menores e reutilizáveis.

- **Definição:** Usamos a palavra-chave `def` seguida do nome da função e parênteses.
- **Parâmetros:** São as variáveis que a função recebe para trabalhar.
- **Retorno (return):** É o resultado que a função envia de volta para quem a chamou. Se não houver `return`, a função executa a tarefa mas "morre" ali mesmo.
```

```python
# Exemplo 3: Função com retorno (return)

def somar(a, b):
    resultado = a + b
    return resultado

def calcular_area_retangulo(largura, altura):
    area = largura * altura
    return area

# Usando as funções
soma = somar(10, 5)
print(f"10 + 5 = {soma}")

area = calcular_area_retangulo(5, 3)
print(f"Área do retângulo (5x3): {area} m²")
```
