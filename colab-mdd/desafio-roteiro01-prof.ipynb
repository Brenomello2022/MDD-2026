```python
Desafio
Contexto
Você foi contratado como analista de dados em uma empresa de mineração. Sua missão é criar um sistema em Python para analisar dados de produção de diferentes minas.

Dados Disponíveis
Você receberá dados sobre:

Nome da mina
Produção diária (em toneladas)
Mineral extraído (Ferro, Ouro, Cobre, Bauxita)
Status operacional (ativa ou inativa)
Objetivos do Desafio
Você deve criar funções e análises que respondam:

Qual a produção total de todas as minas?
Qual mina teve a maior produção?
Quantas minas estão ativas vs inativas?
Qual a produção média por tipo de mineral?
Gerar um relatório completo das minas
Simular variação aleatória na produção (+-10%)
```

```python
# Resposta 2

def encontrar_maior_producao_1(lista_minas):
    quantidade = 0
    for i in lista_minas:
        quantidade += 1

    mina_maior = 0
    controle = 0
    i = 0
    while i < quantidade:
        if lista_minas[i][1] > controle:
            controle = lista_minas[i][1]
            mina_maior = lista_minas[i]
        i += 1

    return mina_maior
encontrar_maior_producao_1(minas)
```

```python
# Resposta 3

def contar_status_minas(lista_minas):
    ativas = 0
    inativas = 0

    for mina in lista_minas:
        if mina[3]:  # Índice 3 = status ativo (True/False)
            ativas += 1
        else:
            inativas += 1

    return (ativas, inativas)
contar_status_minas(minas)
```

```python
# Resposta 4

def calcular_media_por_mineral(lista_minas):

    soma_ferro = 0
    qtd_ferro = 0

    soma_ouro = 0
    qtd_ouro = 0

    soma_cobre = 0
    qtd_cobre = 0

    soma_bauxita = 0
    qtd_bauxita = 0

    for mina in lista_minas:
        producao = mina[1]
        mineral = mina[2]

        if mineral == "Ferro":
            soma_ferro += producao
            qtd_ferro += 1
        elif mineral == "Ouro":
            soma_ouro += producao
            qtd_ouro += 1
        elif mineral == "Cobre":
            soma_cobre += producao
            qtd_cobre += 1
        elif mineral == "Bauxita":
            soma_bauxita += producao
            qtd_bauxita += 1

    if qtd_ferro > 0:
        media_ferro = soma_ferro / qtd_ferro
        print(f"Ferro: {media_ferro} toneladas (minas: {qtd_ferro})")

    if qtd_ouro > 0:
        media_ouro = soma_ouro / qtd_ouro
        print(f"Ouro: {media_ouro} toneladas (minas: {qtd_ouro})")

    if qtd_cobre > 0:
        media_cobre = soma_cobre / qtd_cobre
        print(f"Cobre: {media_cobre} toneladas (minas: {qtd_cobre})")

    if qtd_bauxita > 0:
        media_bauxita = soma_bauxita / qtd_bauxita
        print(f"Bauxita: {media_bauxita} toneladas (minas: {qtd_bauxita})")

calcular_media_por_mineral(minas)
```

```python
# Resposta 5


def gerar_relatorio(lista_minas):

    print("=" * 60)
    print("📋 RELATÓRIO COMPLETO DE MINERAÇÃO")
    print("=" * 60)
    print()

    for i, mina in enumerate(lista_minas, start=1):

        status_texto = "🟢 ATIVA" if mina[3] else "🔴 INATIVA"


        if mina[1] > 800:
            nivel = "ALTA"
        elif mina[1] > 400:
            nivel = "MÉDIA"
        elif mina[1] > 0:
            nivel = "BAIXA"
        else:
            nivel = "PARADA"

        print(f"{i}. {mina[0]}")
        print(f"   Mineral: {mina[2]}")
        print(f"   Produção: {mina[1]} toneladas - Nível: {nivel}")
        print(f"   Status: {status_texto}")
        print()

gerar_relatorio(minas)
```

```python
# Resposta 6

import random
import math

def simular_variacao_producao(lista_minas, percentual_variacao=10):

    novas_minas = []

    for mina in lista_minas:

        if mina[3] and mina[1] > 0:
            variacao = random.uniform(-percentual_variacao, percentual_variacao)
            nova_producao = mina[1] * (1 + variacao / 100)
            nova_producao = math.floor(nova_producao)

            if nova_producao < 0:
                nova_producao = 0

            print(f"📊 {mina[0]}:")
            print(f"   Produção original: {mina[1]} ton")
            print(f"   Variação: {variacao:+.1f}%")
            print(f"   Nova produção: {nova_producao} ton")
            print()

            nova_mina = [mina[0], nova_producao, mina[2], mina[3]]
        else:
            nova_mina = mina.copy()

        novas_minas.append(nova_mina)

    return novas_minas

minas_atualizadas = simular_variacao_producao(minas)
```
