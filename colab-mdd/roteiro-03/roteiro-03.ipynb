Limpeza dos dos dados:

A limpeza geralmente ocorre dentro do processo de ETL (Extract, Transform, Load).
Extract (Extração): Os dados brutos são extraídos das fontes (APIs, Bancos de Produção, Planilhas).

Transform (Transformação): Os dados são levados para uma Staging Area (uma área de rascunho) onde as regras de limpeza e padronização são aplicadas.

Load (Carga): Os dados limpos são inseridos nas tabelas de Fato e Dimensão.

1. Dados de Identificação (Chaves/Keys)
Geralmente são números inteiros que servem para ligar as tabelas.

id, pesid, id_veiculo: Tipo Integer (Inteiro).

Função: São as chaves primárias ou estrangeiras. Elas não servem para fazer contas (você não soma o "id"), mas sim para identificar registros únicos.

2. Dados Temporais (Date/Time)
data_inversa: Tipo Date (ex: 2023-05-20).

horario: Tipo Time (ex: 14:30:00).

dia_semana: Tipo String/Categorical (ex: "segunda-feira").

Dica de Limpeza: Transformar a data_inversa em um objeto de data real permite extrair o mês, ano ou semestre para relatórios.

3. Dados Categóricos (Dimensões - Texto)
Estes são os atributos descritivos. No banco de dados, são armazenados como Varchar ou String.

Geográficos: uf, municipio, regional, delegacia, uop.

Causa e Tipo: causa_principal, causa_acidente, tipo_acidente, classificacao_acidente.

Condições: fase_dia, sentido_via, condicao_metereologica, tipo_pista, tracado_via, uso_solo.

Veículo/Pessoa: tipo_veiculo, marca, tipo_envolvido, estado_fisico, sexo.

4. Dados Numéricos Contínuos e Quantitativos (Fatos - Medidas)
Estes são os campos onde aplicamos operações matemáticas (soma, média, etc.).

ilesos, feridos_leves, feridos_graves, mortos: Tipo Integer. São as métricas centrais da sua "Tabela de Fatos".

idade: Tipo Integer. Pode ser uma métrica (média de idade) ou uma dimensão (faixa etária).

km: Tipo Float (Decimal). Indica a localização exata na via.

latitude, longitude: Tipo Float/Decimal de alta precisão. Essencial para mapas de calor (GIS).

ano_fabricacao_veiculo: Tipo Integer (ano).
