```python
## 3) Comandos de Sistema (Magic Commands)

No ambiente de Notebook, existem comandos especiais que começam com `%` (para a linha) ou `%%` (para a célula inteira) que ajudam a gerenciar o ambiente.

- **%who**: Lista todas as variáveis que você criou na memória até agora.
- **%timeit**: Mede quanto tempo um pedaço de código leva para ser executado (ótimo para performance).
- **!ls** ou **!dir**: Permite executar comandos do terminal do sistema operacional diretamente de uma célula de código.
```

```python
## 4) Descobrindo Tipos de Dados com type()

Como o Python é uma linguagem de **tipagem dinâmica**, às vezes precisamos confirmar que tipo de dado está guardado em uma variável.

**Sintaxe:** `type(variavel)`

**Exemplos Comuns:**
- `<class 'int'>` para números inteiros.
- `<class 'float'>` para números com casas decimais.
- `<class 'str'>` para textos (strings).
- `<class 'bool'>` para valores lógicos (True/False).
```

```python
## 5) Conversão de Tipos (Casting)

Muitas vezes recebemos um dado como texto (ex: vindo de um formulário) e precisamos transformá-lo em número para fazer cálculos.

- **int()**: Converte para inteiro. Ex: `int("10")` vira `10`.
- **float()**: Converte para número decimal. Ex: `float(5)` vira `5.0`.
- **str()**: Converte qualquer valor em texto. Ex: `str(100)` vira `"100"`.

**Dica:** O casting é essencial para evitar erros de tipo ao concatenar textos com números!
```

```python
## 7) Operadores Lógicos

São usados para combinar múltiplas condições em uma única expressão, retornando sempre `True` ou `False`.

- **and** (e): Retorna verdadeiro apenas se **todas** as condições forem verdadeiras.
- **or** (ou): Retorna verdadeiro se **pelo menos uma** das condições for verdadeira.
- **not** (não): Inverte o valor lógico. Se algo é `True`, o `not` transforma em `False`.

**Exemplo:** `(idade >= 18) and (tem_carteira == True)`
```

```python
## 8) Listas (Mutáveis)

As **Listas** são coleções ordenadas de itens que podem ser alterados (mutáveis). No Python, usamos colchetes `[]` para criá-las. No Notebook, elas são exibidas de forma muito organizada.

- **Sintaxe:** `minha_lista = [1, 2, "texto", True]`
- **Acesso por Índice:** O Python começa a contar do 0. Então, `minha_lista[0]` é o primeiro item.
- **Mutabilidade:** Podes mudar um valor específico da lista a qualquer momento.
```

```python
minha_lista = [1, 2, "texto", True]
minha_lista

minha_lista[-1]
```

```python
## 9) Tuplas (Imutáveis)

As **Tuplas** são muito parecidas com as listas, mas com uma diferença crucial: elas são **imutáveis**. Uma vez criadas, não podes alterar, adicionar ou remover itens. Usamos parênteses `()` para criá-las.

- **Sintaxe:** `minha_tupla = (10, 20, 30)`
- **Por que usar?** São mais rápidas que as listas e garantem que dados sensíveis (como coordenadas geográficas ou configurações) não sejam alterados acidentalmente pelo código.
```

```python
minha_tupla = (10, 20, 30)
minha_tupla

minha_tupla = (10, 20, 30)
minha_tupla[0]

minha_tupla
# desempacotar
10, 20, 30 == minha_tupla
```

```python
## 10) Estruturas Condicionais (if, elif, else)

As estruturas condicionais permitem que o código siga caminhos diferentes dependendo de uma condição lógica.

**Importante:** Em Python, a **indentação** (o espaço no início da linha) é obrigatória. É ela que diz ao Python o que está "dentro" do seu `if`.

- **if:** "Se" a condição for verdadeira, faça isso.
- **elif:** "Senão, se" a condição anterior foi falsa, tente esta.
- **else:** "Senão", se nada acima funcionou, faça isso por padrão.
```

```python
usuario = 'admin'
senha = '12345'

if usuario == 'admin' and senha == '1234':
  print('Bem vindo, Admin!')
else:
  print('Recusado')
```

```python
## 11) Estruturas de Repetição (for e while)

As estruturas de repetição permitem que você execute o mesmo bloco de código várias vezes, poupando linhas de programação.

### for (para)
É usado quando você sabe (ou o Python sabe) exatamente quantas vezes repetir.
- **Percorrer listas:** Ele pega um item por vez da lista até chegar ao último.
- **Função range():** Cria uma sequência numérica (ex: `range(5)` gera de 0 a 4) para repetições controladas.

### while (enquanto)
É usado quando você não sabe exatamente quantas vezes o código vai rodar. Ele continuará repetindo enquanto uma condição for verdadeira.

⚠️ **Cuidado com o Loop Infinito:** Se a condição do `while` nunca se tornar falsa, o código nunca para de rodar. É necessário garantir que algo mude dentro do bloco para encerrar a repetição.
```

```python
for i in range(0, 5):
  print(i)
```

```python
## 12) Importando Bibliotecas (import)

Uma **biblioteca** é um conjunto de códigos escritos por outras pessoas que você pode usar no seu projeto para não precisar "reinventar a roda". A biblioteca `random`, por exemplo, serve para gerar números aleatórios ou fazer escolhas randômicas.

- **import:** Carrega a biblioteca inteira para o seu Notebook.
- **Uso:** Após importar, você acessa as funções usando o ponto (Ex: `biblioteca.funcao()`).

Algumas bibliotecas populares:
- `random` - números aleatórios
- `math` - funções matemáticas avançadas
- `datetime` - trabalhar com datas e horas
```

```python
## 13) Funções (def)

Uma **função** é um bloco de código que só é executado quando é chamado. Elas servem para evitar a repetição de código e para organizar a lógica em partes menores e reutilizáveis.

- **Definição:** Usamos a palavra-chave `def` seguida do nome da função e parênteses.
- **Parâmetros:** São as variáveis que a função recebe para trabalhar.
- **Retorno (return):** É o resultado que a função envia de volta para quem a chamou. Se não houver `return`, a função executa a tarefa mas "morre" ali mesmo.
```

```python
def _soma(a, b):
  return a+b
```
